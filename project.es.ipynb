#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
#comparativa de precios promedios entre Valdemorrillo y Galapagar
Promed_Valdemorillo = (ds[ds["level5"] == "Valdemorillo"]["price"].sum()/ds[ds["level5"] == "Valdemorillo"]["price"].count())*100
Promed_Galapagar = (ds[ds["level5"] == "Galapagar"]["price"].sum()/ds[ds["level5"] == "Galapagar"]["price"].count())*100
print(Promed_Valdemorillo)
print(Promed_Galapagar)
```

Se observa una diferencia aunque mínima, existente, por lo que se podría estimar sin ser concluyentes, que podría haber algun rango de similitud de los precios de las viviendas en ambas locaciones, probablemente por compatibilidad demográfica

#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (precio/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio por metro cuadrado y escribe una conclusión sobre ellos.

Pista: Crea una nueva columna llamada `pps` (*price per square* o precio por metro cuadrado) y luego analiza los valores.

```python
# calcular precios por metro cuadrado, añadir nueva columna
metr_cuad_tot = ds.groupby('level5')['surface'].transform('sum')
precio_total = ds.groupby('level5')['price'].transform('sum')
ds['PPS'] = (precio_total / metr_cuad_tot)*100
ds
```

```python
Sum_precio_vald = ds[ds["level5"] == "Valdemorillo"]["price"].sum()
Sum_metr2_vald = ds[ds["level5"] == "Valdemorillo"]["surface"].sum()
Prom_pps_vald = (Sum_precio_vald / Sum_metr2_vald)*100

Sum_precio_Gala = ds[ds["level5"] == "Galapagar"]["price"].sum()
Sum_metr2_Gala = ds[ds["level5"] == "Galapagar"]["surface"].sum()
Prom_pps_Gala = (Sum_precio_Gala / Sum_metr2_Gala)*100

print(Prom_pps_vald)
print(Prom_pps_Gala)
```

Al comparar precios promedio por metro cuadrado en ambos sectores, se observa que no existe una diferencia significativa, por lo que se estima una similitud geo-económica entre ambas localidades
